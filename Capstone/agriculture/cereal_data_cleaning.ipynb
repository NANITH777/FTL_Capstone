```python
import pandas as pd
import os
# Spécifiez le chemin du répertoire contenant les fichiers CSV
directory = r'D:\Capstone\Countries'

# Obtenez la liste de tous les fichiers CSV dans le répertoire
files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Parcourez chaque fichier CSV
for file in files:
    file_path = os.path.join(directory, file)
    
    # Chargez le fichier CSV dans un DataFrame
    df = pd.read_csv(file_path)
    
    # Supprimez la deuxième ligne (index 1) du DataFrame
    df = df.drop(0).reset_index(drop=True)
    
    # Enregistrez le DataFrame modifié dans le même fichier CSV
    df.to_csv(file_path, index=False)
    
    print(f'Second line removed from {file}')
```

```python
directory = r'D:\Capstone\Countries'

features = [
    'CO2 emissions (kt)',
    'Other greenhouse gas emissions, HFC, PFC and SF6 (thousand metric tons of CO2 equivalent)',
    'Methane emissions (kt of CO2 equivalent)',  
    'Average precipitation in depth (mm per year)',
    'Droughts, floods, extreme temperatures (% of population affected)',
    'Annual freshwater withdrawals, total (% of internal resources)',
    'Cereal yield (kg per hectare)'
]

all_dataframes = []

files = [f for f in os.listdir(directory) if f.endswith('.csv')]

for file in files:
    file_path = os.path.join(directory, file)
    
    df = pd.read_csv(file_path)
    
    if 'Indicator Name' in df.columns:
        filtered_df = df[df['Indicator Name'].isin(features)]
        
        all_dataframes.append(filtered_df)
        
        print(f'Processed {file}')
    else:
        print(f'Column "Indicator Name" not found in {file}')

if all_dataframes:
    merged_df = pd.concat(all_dataframes, ignore_index=True)

    output_file = r'D:\Capstone\agriculture\cereal.csv'
    merged_df.to_csv(output_file, index=False)

    print(f'Merged data saved to {output_file}')
else:
    print('No valid data to merge.')
```

```python
df = pd.read_csv("cereal.csv")
df.head()
```

```python
df.drop(columns=["Country ISO3", "Indicator Code"], inplace=True)
df.head()
```

```python
# Use the pivot_table method to convert "Indicator Name" into columns
df_pivoted = df.pivot_table(index=['Country Name', 'Year'], columns='Indicator Name', values='Value')

# Reset the index to make "Country Name" and "Year" regular columns
df_pivoted = df_pivoted.reset_index()

# Define a dictionary to map the original names to the brief names
renamed_columns = {
    'CO2 emissions (kt)': 'CO2 emissions (kt)',
    'Other greenhouse gas emissions, HFC, PFC and SF6 (thousand metric tons of CO2 equivalent)': 'Other GHG emissions (thousand tons)',
    'Methane emissions (kt of CO2 equivalent)': 'Methane emissions (kt CO2 eq.)',
    'Average precipitation in depth (mm per year)': 'Avg precipitation (mm/year)',
    'Droughts, floods, extreme temperatures (% of population affected)': 'Extreme weather (% pop. affected)',
    'Annual freshwater withdrawals, total (% of internal resources)': 'Freshwater withdrawals (% internal)',
    'Cereal yield (kg per hectare)': 'Cereal yield (kg/hectare)'
}

# Rename the columns
df_pivoted = df_pivoted.rename(columns=renamed_columns)

# Display the first few rows of the renamed DataFrame
df_pivoted.tail()
```

```python
# Save the pivoted DataFrame to a CSV file named cerealmix.csv
output_file = r'D:\Capstone\agriculture\cerealmixed.csv'
df_pivoted.to_csv(output_file, index=False)

print(f'Pivoted data saved to {output_file}')
```
